```python
#TASK_1
import math

class Student:
    ID = 0

    def __init__(self, name, department, age, cgpa):
        self.name = name
        self.dept = department
        self.age = age
        self.cgpa = cgpa

    def get_details(self):
        Student.ID += 1
        print("ID:", Student.ID)
        print("Name:", self.name)
        print("Department:", self.dept)
        print("Age:", self.age)
        print("CGPA:", self.cgpa)

    @classmethod
    def from_string(cls, info):
        name, department, age, cgpa = info.split('-')
        obj = cls(name, department, age, cgpa)
        return obj


s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_string("Sumaiya-BBA-23-3.96")
s4.get_details()
print("-----------------------")
print("Subtask 5: Difference between a class variable and an instance varibale is, instance variable is dependent on object but class variable is not dependent on object")

print("subtask 6: Difference between a class method and an instance method is, class methods works on the same class and cannot access instance variable but instance method works on object and can access on instance variable")
```

```python
#TASK_2
class Dates:
    def __init__(self, date):
        self.date = date

    def getDate(self):
        return self.date

    @classmethod
    def toDashDate(cls, ano_date):
        date = ano_date.replace("/", "-")
        obj = cls(date)
        return obj


date1 = Dates("05-09-2020")
dateFromDB = "05/09/2020"
date2 = Dates.toDashDate(dateFromDB)
if(date1.getDate() == date2.getDate()):
    print("Equal")
else:
    print("Unequal")
print("SubTask 4: ", "In the conditional statement it prints 'Equal'.Because, At first dateFromDB was in slash mode,Then"
      " after calling the method toDashDate the slashes became Dashes and stores in date2, date1 is already in dash mode. "
      "So now date1 and date2 stores  the same date with dashes.That's why it prints equal.")
```

```python
#TASK_3
class Passenger:
    count = 0

    def __init__(self, name):
        self.name = name
        Passenger.count += 1

    def set_bag_weight(self, weight):
        self.weight = weight
        bus_fair = 450
        if self.weight <= 20:
            obj = bus_fair
            self.obj = obj
        elif self.weight >= 21 and self.weight <= 50:
            obj = bus_fair + 50
            self.obj = obj
        else:
            obj = bus_fair + 100
            self.obj = obj

    def printDetail(self):
        print("Name:", self.name)
        print("Bus Fare:", self.obj)


print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
# TASK 4
class Travel:
    count = 0

    def __init__(self, source, destination):
        self.source = source
        self.destination = destination
        self.time = 1
        Travel.count += 1

    def set_destination(self, new_d):
        self.new_d = new_d
        self.destination = self.new_d

    def set_source(self, new_s):
        self.new_s = new_s
        self.source = self.new_s

    def set_time(self, flight_time):
        self.time = flight_time

    def display_travel_info(self):
        obj = 'Source:' + self.source + '\nDestination:' + \
            self.destination + '\nFlight Time:' + str(self.time) + ':00'
        return obj

print("No. of Traveller =", Travel.count)
print("=======================")
t1 = Travel("Dhaka", "India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur", "Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka", "New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka", "India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print("No. of Traveller =", Travel.count)
```

```python
#TASK_5
class Fruit:
    fruitCount = 0

    def __init__(self, name, count):
        self.name = name
        self.count = count
        Fruit.fruitCount = self.count + Fruit.fruitCount

    @staticmethod
    def saySomethingGood():
        print("Fruits are good for health")

    @staticmethod
    def resetcount():
        Fruit.fruitCount = 0


apples = Fruit("Apple", 3)
pears = Fruit("Pear", 4)
print(apples.name, apples.count)
print(pears.name, pears.count)
print("Total number of fruits", Fruit.fruitCount)
Fruit.saySomethingGood()
Fruit.resetcount()
print("Total number of fruits", Fruit.fruitCount)
```

```python
# TASK 6
class Cat:
    Number_of_cats = 0
    action = "sitting"
    color = "white"

    def __init__(self, color, action):
        self.color = color
        self.action = action
        Cat.Number_of_cats += 1

    @classmethod
    def no_parameter(cls):
        obj1 = cls(cls.color, cls.action)
        return obj1

    @classmethod
    def first_parameter(cls, color):
        cls.color = color
        obj2 = cls(cls.color, cls.action)
        return obj2

    @classmethod
    def second_parameter(cls, action):
        cls.action = action
        obj3 = cls("Grey", cls.action)
        return obj3

    def printCat(self):
        print(self.color, " cat is ", self.action)

    def changeColor(self, new_C):
        self.color = new_C
        
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#TASK_7
class Cylinder:
    radius = 5
    height = 18

    def __init__(self, r, h):
        self.radius = r
        self.height = h
        print("Default radius", str(Cylinder.radius),
              "and height", str(Cylinder.height))
        Cylinder.radius = r
        Cylinder.height = h
        print("Updated radius", Cylinder.radius, "and height", Cylinder.height)

    @classmethod
    def swap(cls, radius, height):
        obj = cls(height, radius)
        return obj

    @classmethod
    def changeFormat(cls, all):
        new_r, new_h = all.split("-")
        obji = cls(float(new_r), float(new_h))
        return obji

    @staticmethod
    def volume(r, h):
        print("volume: "+str(math.pi*r*r*h))

    @staticmethod
    def area(r, h):
        print("Area: "+str((2*math.pi*r*r)+(2*math.pi*r*h)))


c1 = Cylinder(0, 0)
Cylinder.area(c1.radius, c1.height)
Cylinder.volume(c1.radius, c1.height)
print("===============================")
c2 = Cylinder.swap(8, 3)
c2.area(c2.radius, c2.height)
c2.volume(c2.radius, c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius, c3.height)
c3.volume(c3.radius, c3.height)
print("===============================")
Cylinder(0.3, 5.56).area(Cylinder.radius, Cylinder.height)
print("===============================")
Cylinder(3, 5).volume(Cylinder.radius, Cylinder.height)
```
