```python
# task 1
number_1 = int(input('Enter a number: '))
number_2 = int(input('Enter another number: '))


def Fraction_function(number_1, number_2):

    if number_1 != 0 and number_2 != 0 and (number_1 % number_2) != 0:
        division = number_1 / number_2
        return (division - int(division))
    else:
        return 0

print(Fraction_function(number_1, number_2))
```

```python
# TASK 2
def bmi_score(height, weight):
    height = (height/100)**2
    bmi = weight/height
    if bmi < 18.5:
        print("Score is", bmi, "You are Underweight")
    if bmi >= 18.5 and bmi <= 24.9:
        print("Score is", bmi, "You are Normal")
    if bmi >= 25 and bmi <= 30:
        print("Score is", bmi, "You are Overweight")
    if bmi > 30:
        print("Score is", bmi, "You are Obese")


bmi_score(177, 67)
```

```python
# TASK 3
def min_max_div(minimum, maximum, divisor):
    total = 0
    for x in range(minimum, maximum):
        if x % divisor == 0:
            total = total + x
    return total

print(min_max_div(3, 16, 3))
```

```python
# TASK 4
def order(burger, place="Mohakhali"):
    if place == "Mohakhali":
        delivery_charge = 40
    else:
        delivery_charge = 60
    if burger == "BBQ Chicken Cheese Burger":
        meal_cost = 250
    elif burger == "Beef Burger":
        meal_cost = 170
    elif burger == "Naga Drums":
        meal_cost = 200
    tax = meal_cost*(8/100)
    total_price = meal_cost+delivery_charge+tax
    print(total_price)

order("Beef Burger", "Dhanmondi")
```

```python
# task_5
def check(email, new_domain, old_domain="kaaj.com"):
    if old_domain in email:
        index = email.index("@")
        final_mail = "Changed:" + email[:index + 1:] + new_domain
    else:
        final_mail = "Unchanged:" + email
    print(final_mail)

check("alice@kaaj.com", "sheba.xyz", "kaaj.com")
```

```python
# task 6
def find_vowels(full_name):
    count = 0
    vowels = ""
    string = ['a', 'e', 'i', 'o', 'u']
    for char in full_name:
        if char in string:
            count += 1
            vowels += char + ","
    print("Vowels:", vowels[0:-1], "Total number of vowels:", count)

find_vowels("Steve Jobs")
```

```python
# TASK 7
def palindrome(string):
    string = string.replace(" ", "")
    if string == string[::-1]:
        print("Palindrome")
    else:
        print("Not a palindrome")

palindrome("hello")
```

```python
# TASK 8
def dayconverter(days):
    years = days//365
    months = (days % 365)//30
    days = (days % 365) % 30
    print(years, "years", months, "months", days, "days")


dayconverter(int(input("Enter a number:")))
```

```python
# task_9
def teacher(sentence):
    new_sentence = ""
    word1 = sentence[0].upper()
    new_sentence += word1
    for count in range(1, len(sentence)):
        if (sentence[count - 2] == "." or sentence[count - 2] == "!" or sentence[count - 2] == "?") and sentence[count-1] == " ":
            word = sentence[count].upper()
            new_sentence += word
        elif sentence[count] == 'i' and sentence[count - 1] == " " and sentence[count + 1] == ' ':
            word = sentence[count].upper()
            new_sentence += word
        else:
            new_sentence += sentence[count]

    return new_sentence

print(teacher('my favourite animal is a dog. a dog has sharp teeth so that it can eat flesh very easily. do you know my pet dog’s name? i love my pet very much.'))
```
