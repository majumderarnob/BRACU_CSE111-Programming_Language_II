                                                  String Problems:

```python
# TASK 1
user_input = input("Enter a string: ")
upperCase = 0
lowerCase = 0
for i in range(len(user_input)):
    if user_input[i].isupper():
        upperCase += 1
    if user_input[i].islower():
        lowerCase += 1
for j in range(len(user_input)):
    if upperCase > lowerCase:
        print(user_input.upper())
        break
    elif lowerCase >= upperCase:
        print(user_input.lower())
        break
```

```python
# TASK 2
user_input = input("Enter a string: ")
number_type = True
word_type = True
for i in range(len(user_input)):
    if ord(user_input[i]) >= 48 and ord(user_input[i]) <= 57:
        word_type = False
    if ord(user_input[i]) >= 65 and ord(user_input[i]) <= 90 or ord(user_input[i]) >= 97 and ord(user_input[i]) <= 122:
        number_type = False
if number_type == True:
    print("NUMBER")
elif word_type == True:
    print("WORD")
else:
    print("MIXED")
```

```python
# task_6
password = input('Enter your password: ')
special_char = ['_', '$', '#', '@']
upperCase = 0
lowerCase = 0
specialCase = 0
digit = 0

for char in password:
    if chr(65) <= char <= chr(91):
        upperCase += 1
    if chr(97) <= char <= chr(123):
        lowerCase += 1
    if chr(48) <= char <= chr(57):
        digit += 1
    if char in special_char:
        specialCase += 1

if lowerCase < 1:
    print('Lowercase Missing', end=',')
if upperCase < 1:
    print('Uppercase Missing', end=',')
if digit < 1:
    print('Digit Missing', end=',')
if specialCase < 1:
    print('Special Character Missing')
if lowerCase > 0 and upperCase > 0 and digit > 0 and specialCase > 0:
    print('OK')
```

```python
#task_1
list1 = []
while True:
    num = input("Enter a Number: ")
    if num == "STOP":
        break
    else:
        num = int(num)
        list1.append(num)
list2 = []
for x in list1:
    if x not in list2:
        list2.append(x)
for x in list2:
    count = 0
    for y in list1:

        if x == y:
            count += 1
    print(x, "-", count, "times")
```

```python
# task_5
user_input = input("Enter a string: ")
lowerCase = []
upperCase = []
num = []
oddNum = []
evenNum = []
const1 = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
const2 = "abcdefghijklmnopqrstuvwxyz"
const3 = "1234567890"

for x in user_input:
    if x in const2:
        lowerCase.append(x)
    if x in const1:
        upperCase.append(x)
    if x in const3:
        num.append(x)

for y in num:
    if int(y) % 2 == 0:
        evenNum.append(y)
    else:
        oddNum.append(y)

lowerCase.sort()
upperCase.sort()
oddNum.sort()
evenNum.sort()
output = lowerCase + upperCase + oddNum + evenNum
for x in output:
    print(x, end="")
```

                                  Dictionary & Tuple Problems

```python
# task_1
dict1 = {'a': 100, 'b': 100, 'c': 200, 'd': 300}
dict2 = {'a': 300, 'b': 200, 'd': 400, 'e': 200}
new_dict = {}
for key, value in dict1.items():
    if key in dict2:
        new_dict[key] = dict1[key] + dict2[key]
    else:
        new_dict[key] = value
for key, value in dict2.items():
    if key not in dict1:
        new_dict[key] = value
print(new_dict)

create = []
for key, value in new_dict.items():
    if value not in create:
        create.append(new_dict[key])

create.sort()
unique_value = tuple(create)
print(unique_value)
```
