```python
# Task_1
class Calculator:
    def __init__(self, first, operator, last):
        self.first_value = first
        self.operator = operator
        self.last_value = last
        print('Let\'s Calculate!')
        print(f'Value 1: ', self.first_value)
        print(f'Operator: ', self.operator)
        print(f'Value 2: ', self.last_value)

    def add(self):
        output = self.first_value + self.last_value
        return f'Result: {output}'

    def subtract(self):
        output = self.first_value - self.last_value
        return f'Result: {output}'

    def multiply(self):
        output = self.first_value * self.last_value
        return f'Result: {output}'

    def divide(self):
        output = self.first_value / self.last_value
        return f'Result: {output}'

val1 = int(input('Enter first value: '))
oprt = input('Enter operator: ')
val2 = int(input('Enter second value: '))
my_cal = Calculator(val1, oprt, val2)
if oprt == '+':
    result = my_cal.add()
elif oprt == '-':
    result = my_cal.subtract()
elif oprt == '*':
    result = my_cal.multiply()
elif oprt == '/':
    result = my_cal.divide()
print(result)
```

```python
# Task_2
class Customer:
    def __init__(self, name):
        self.name = name

    def greet(self, name=None):
        if name == None:
            print('Hello!')
        else:
            print(f'Hello {self.name}!')

    def purchase(self, *args):
        count = 0
        for x in args:
            count += 1
        print(f'{self.name}, you purchased {count} item(s): ')
        for x in args:
            print(x)

customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
# Task_3
class Panda:
    def __init__(self, name, gender, age):
        self.name = name
        self.gender = gender
        self.age = age

    def sleep(self, hours=0):
        if 3 <= hours <= 5:
            diet = 'Mixed Vegies'
        elif 6 <= hours <= 8:
            diet = 'Eggplant & Tofu'
        elif 9 <= hours <= 11:
            diet = 'Broccoli Chicken'
        else:
            diet = 'bamboo leaves'
        if hours == 0:
            return f'{self.name}\'s duration is unknown thus should have only {diet}'
        else:
            return f'{self.name} sleeps {hours} hours daily and should have {diet}'


panda1 = Panda("Kunfu", "Male", 5)
panda2 = Panda("Pan Pan", "Female", 3)
panda3 = Panda("Ming Ming", "Female", 8)
print("{} is a {} Panda Bear who is {} years old".format(
    panda1.name, panda1.gender, panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(
    panda2.name, panda2.gender, panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(
    panda3.name, panda3.gender, panda3.age))

print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
# Task_4
class Cat:
    def __init__(self, color='White', action='sitting'):
        self.color = color
        self.action = action

    def printCat(self):
        print(f'{self.color} cat is {self.action}')

    def changeColor(self, color):
        self.color = color

c1 = Cat()
c2 = Cat("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
```

```python
# Task_5
class Vehicle:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def moveUp(self):
        self.y += 1

    def moveDown(self):
        self.y -= 1

    def moveRight(self):
        self.x += 1

    def moveLeft(self):
        self.x -= 1

    def print_position(self):
        print(f'({self.x},{self.y})')

car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
# Task_6
class Programmer:
    def __init__(self, name, language, experience):
        self.name = name
        self.lan = language
        self.exp = experience
        print('Horray! A new programmer is born')

    def addExp(self, n=0):
        self.exp += n
        print(f'Updating experience of {self.name}')

    def printDetails(self):

        print(f'Name: {self.name}')
        print(f'Language: {self.lan}')
        print(f'Experience: {self.exp}')


p1 = Programmer("Ethen Hunt", "Java", 10)
p1.printDetails()
print('--------------------------')
p2 = Programmer("James Bond", "C++", 7)
p2.printDetails()
print('--------------------------')
p3 = Programmer("Jon Snow", "Python", 4)
p3.printDetails()
p3.addExp(5)
p3.printDetails()
```

```python
# Task_7
class Student:
    def __init__(self, name, id, department='CSE'):
        self.name = name
        self.id = id
        self.dept = department

    def dailyEffort(self, ef=None):
        self.effort = ef
        if ef <= 2:
            sugs = 'Should give more effort!'
        elif ef <= 4:
            sugs = 'Keep up the good work!'
        else:
            sugs = 'Excellent! Now motivate others.'
        self.suggestion = sugs

    def printDetails(self):
        print(f'Name: {self.name}')
        print(f'ID: {self.id}')
        print(f'Department: {self.dept}')
        print(f'Daily Effort: {self.effort} hour(s)')
        print(f'Suggestion: {self.suggestion}')


harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
# Task_8
class Patient:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def add_Symptom(self, *args):
        lst1 = []
        for x in args:
            lst1.append(x)
        symp = ', '.join(lst1)
        self.symp = symp

    def printPatientDetail(self):
        print(f'Name: {self.name}')
        print(f'Age: {self.age}')
        print('Symptoms: ', self.symp)

p1 = Patient('Thomas', 23)
p1.add_Symptom('Headache')
p2 = Patient('Carol', 20)
p2.add_Symptom('Vomiting', 'Coughing')
p3 = Patient('Mike', 25)
p3.add_Symptom('Fever', 'Headache', 'Coughing')
print('=========================')
p1.printPatientDetail()
print('=========================')
p2.printPatientDetail()
print('=========================')
p3.printPatientDetail()
print('=========================')
```

```python
# Taks_9
class Avengers:
    def __init__(self, name, partner):
        self.name = name
        self.partner = partner

    def super_powers(self, *args):
        lst1 = []
        for x in args:
            lst1.append(x)
        power = ', '.join(lst1)
        self.power = power

    def printAvengersDetail(self):
        print(f'Name: {self.name}')
        print(f'Partner: {self.partner}')
        print(f'Super powers: {self.power}')

a1 = Avengers('Captain America', 'Bucky Barnes')
a1.super_powers('Stamina', 'Slowed ageing')
a2 = Avengers('Doctor Strange', 'Ancient One')
a2.super_powers('Mastery of magic')
a3 = Avengers('Iron Man', 'War Machine')
a3.super_powers('Genius level intellect', 'Scientist ')
print('=========================')
a1.printAvengersDetail()
print('=========================')
a2.printAvengersDetail()
print('=========================')
a3.printAvengersDetail()
print('=========================')
```

```python
# Task_10
class Shinobi:
    def __init__(self, name, rank, salary=0, mission=0):
        self.name = name
        self.rank = rank
        self.salary = salary
        self.mission = mission

    def changeRank(self, rank):
        self.rank = rank

    def calSalary(self, mission):
        self.mission = mission
        if self.rank == 'Genin':
            self.salary = self.mission * 50
        elif self.rank == 'Chunin':
            self.salary = self.mission * 100
        else:
            self.salary = self.mission * 500

    def printInfo(self):
        print(f'Name: {self.name}')
        print(f'Rank: {self.rank}')
        print(f'Number of mission: {self.mission}')
        print(f'Salary: {self.salary}')

naruto = Shinobi('Naruto', 'Genin')
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', 'Genin')
shikamaru.printInfo()
shikamaru.changeRank('Chunin')
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi('Neiji', 'Jonin')
neiji.calSalary(5)
neiji.printInfo()
```
